# EDA Notebook
Exploration of cleaned NLDC/SCADA hourly demand data.


```python
import sys, os
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
from config import CLEANED_DATA_DIR

df = pd.read_parquet(CLEANED_DATA_DIR / 'energy_cleaned.parquet')
df = df.set_index('datetime').sort_index()
df.head()
```

## Basic shape and summary stats

```python
df.describe()
```

## Missingness check
Cleaned data should have zero nulls (already interpolated), but confirm.

```python
df.isna().sum()
```

## Full time series

```python
fig, ax = plt.subplots(figsize=(14, 4))
df['demand_mw'].plot(ax=ax)
ax.set_title('Hourly demand (MW) over full history')
ax.set_ylabel('MW')
plt.show()
```

## Daily seasonality
Average demand by hour of day.

```python
hourly_avg = df.groupby(df.index.hour)['demand_mw'].mean()
hourly_avg.plot(kind='bar', figsize=(10, 4), title='Average demand by hour of day')
plt.ylabel('MW')
plt.show()
```

## Weekly seasonality
Weekday vs weekend average demand.

```python
df['day_of_week'] = df.index.dayofweek
dow_avg = df.groupby('day_of_week')['demand_mw'].mean()
dow_avg.plot(kind='bar', figsize=(8, 4), title='Average demand by day of week (0=Mon)')
plt.ylabel('MW')
plt.show()
```

## Yearly / long-term trend
Monthly average demand across the full date range, to check for trend or structural breaks.

```python
monthly_avg = df['demand_mw'].resample('MS').mean()
monthly_avg.plot(figsize=(14, 4), title='Monthly average demand')
plt.ylabel('MW')
plt.show()
```

## Distribution
Histogram of hourly demand values, to spot skew or outliers.

```python
df['demand_mw'].hist(bins=50, figsize=(8, 4))
plt.title('Distribution of hourly demand')
plt.xlabel('MW')
plt.show()
```
